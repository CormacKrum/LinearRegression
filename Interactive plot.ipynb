```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X=(pd.read_csv('Linear_X_Train.csv')).values
y=(pd.read_csv("Linear_Y_Train.csv")).values
```

```python
theta_list = np.load("ThetaList.npy")
T0 = theta_list[:,0]
T1 = theta_list[:,1]
```

```python
plt.ion()
```

```python
for i in range(0,50,3):
    y_pred = T1[i]*X+T0[i]
    plt.scatter(X,y)
    plt.plot(X,y_pred,"red")
    plt.draw()
    plt.pause(1)
    plt.clf()
```
