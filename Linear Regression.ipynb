```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("seaborn")
```

```python
X = pd.read_csv("Linear_X_Train.csv").values
y = pd.read_csv("Linear_Y_Train.csv").values

X = (X - X.mean()) / X.std()
```

```python
plt.scatter(X,y)
plt.ylabel("Performance")
plt.xlabel("Hardwork")
plt.show()
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
theta=[0,0]
theta[0] = model.intercept_
theta[1] = model.intercept_
```

```python
X_test = pd.read_csv("Linear_X_Test.csv").values
```

```python
y_pred = model.predict(X_test)
```

```python
plt.figure()
plt.scatter(X,y,label="Training Data",color="green")
plt.ylabel("Performance")
plt.xlabel("Hardwork")
plt.plot(X_test,y_pred,label="Predicted Line",color="red")
plt.legend()
plt.show()
```
