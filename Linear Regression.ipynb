```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("seaborn")
```

```python
X = pd.read_csv("Linear_X_Train.csv").values
y = pd.read_csv("Linear_Y_Train.csv").values

X = (X - X.mean()) / X.std()
```

```python
plt.scatter(X,y)
plt.ylabel("Performance")
plt.xlabel("Hardwork")
plt.show()
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
intercept = model.intercept_
slope = model.coef_
print("y = "+str(slope[0][0])+" x + "+str(intercept[0]))
```

```python
y_pred = model.predict(X)
```

```python
plt.figure()
plt.scatter(X,y,label="Training Data",color="green")
plt.ylabel("Performance")
plt.xlabel("Hardwork")
plt.plot(X,y_pred,label="Predicted Line",color="red")
plt.legend()
plt.show()
```

```python
model.score(X,y)*100
```
