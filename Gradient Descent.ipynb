```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
X = np.arange(11)
y = (X-5)**2
```

```python
plt.plot(X,y)
plt.ylabel("F(X)")
plt.xlabel("X")
plt.show()
```

```python
plt.figure()
plt.plot(X,y)
x=0
lr=0.1
error = []
for i in range(50):
    grad = 2*(x-5)
    x=x-lr*grad
    y_err = (x-5)**2
    error.append(y_err)
    plt.scatter(x,y_err)
```

```python
plt.plot(error)
```
