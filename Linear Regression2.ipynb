```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("seaborn")
```

```python
X = pd.read_csv("Linear_X_Train.csv").values
y = pd.read_csv("Linear_Y_Train.csv").values

X = (X - X.mean()) / X.std()
```

```python
plt.scatter(X,y)
plt.ylabel("Performance")
plt.xlabel("Hardwork")
plt.show()
```

```python
sigmaX = X.sum()
sigmay = y.sum()
Xy = X*y
sigmaXy = Xy.sum()
XX = X**2
sigmaXX = XX.sum()
n = X.shape[0]
slope = (sigmaXy - sigmaX*sigmay)/(sigmaXX - sigmaX**2)
intercept = (sigmay - slope * sigmaX)/n
print("y = "+str(slope)+" x + "+str(intercept))
```

```python
def predict(X):
    return slope*X + intercept
def R2_Score(y,y_pred):
    num = np.sum((y-y_pred)**2)
    den = np.sum((y-y.mean())**2)
    
    return (1-num/den)*100

y_pred = predict(X)
```

```python
plt.figure()
plt.scatter(X,y,label="Training Data",color="green")
plt.ylabel("Performance")
plt.xlabel("Hardwork")
plt.plot(X,y_pred,label="Predicted Line",color="red")
plt.legend()
plt.show()
```

```python
score = R2_Score(y,y_pred)
print(score)
```
