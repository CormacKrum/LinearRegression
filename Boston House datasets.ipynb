```python
one = np.ones((X.shape[0],1))
X = np.hstack((one,X))
```

```python
def hypothesis(x,theta):
    y_pred=0.0
    for i in range(x.shape[0]):
        y_pred += (theta[i] * x[i])
    return y_pred
def gradient_descent(X,y,lr=0.1,max_epoch=300):
    m,n = X.shape
    theta = np.zeros((n,))
    error_list=[]
    for i in range(max_epoch):
        e = error(X,y,theta)
        error_list.append(e)
        grad = gradient(X,y,theta)
        for j in range(n):
            theta[j] -= lr * grad[j]
    return theta,error_list
def error(X,y,theta):
    e=0.0
    m=X.shape[0]
    for i in range(m):
        y_pred = hypothesis(X[i],theta)
        e+=(y[i]-y_pred)**2
    return e/m

def gradient(X,y,theta):
    m,n =X.shape
    grad = np.zeros((n,))
    for j in range(n):
        for i in range(m):
            y_pred = hypothesis(X[i], theta=theta)
            grad[j] += (y_pred - y[i] ) * X[i][j]
    return grad/m
```

```python
theta,error_list=gradient_descent(X,y)
```

```python
def R2_Score(y,y_pred):
    num = np.sum((y-y_pred)**2)
    den = np.sum((y-y.mean())**2)
    return (1-num/den)*100
```

# Optimining Using Vectorization

```python
def hypothesis(X,theta):
    return np.dot(X,theta)

def error(X,y,theta):
    y_pred = hypothesis(X,theta)
    error=np.sum((y-y_pred)**2)
    return error/X.shape[0]

def gradient(X,y,theta):
    y_pred=hypothesis(X,theta)
    grad = np.dot(X.T,(y_pred-y))
    return grad/X.shape[0]

def gradient_descent(X,y,lr=0.1,max_epoch=300):
    theta = np.zeros((X.shape[1],))
    error_list=[]
    for i in range(max_epoch):
        e = error(X,y,theta)
        error_list.append(e)
        
        grad = gradient(X,y,theta)
        theta-=lr*grad
    return theta,error_list
```

```python
theta,error_list=gradient_descent(X,y)
```
